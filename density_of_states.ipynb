#### 2.2 Statistical Distributions

The most generic form to write the density of states $W$ is using a multinomial coefficient, where number of possible partitions can be written as a function of the amount of some countable instance $N$, and each of the $i$ groups they can be distributed among. Hence $W = \frac{N!}{\prod_i N_i!}$, where $N_i$ is generic for some number of particles composing $N$. For instance, imagine two possible states for two particles, $A$ and $B$ for the distinguishable case, and $A=B$ for the indistinguishable case. Below is the table showing the possible states for the distinguishable case:

| $E_1$ | $E_2$ |
|----------|----------|
| $AB$       |          |
| $A$        | $B$        |
| $B$        | $A$        |
|          | $AB$       |

Distinguishable particle distributions are related to Boltzmann's distribution. Let $g$ be the number of available states and $N$ the amount of particles. The number of possible distributions (Boltzmann) $W_B$ would be $W_B = g^N = 2^2 = 4$, as shown in the former table. It is the number of possible permutations from probabilistics, and that is why it is possible to derive the same relation from the previous multimodal coefficients count. Let $g_i^{N_i}$ be the occupation of state $i$ given the degenerancy $g_i$, once there are $N_i$ ways to redistribute these particles without altering the system's energy. This is why $W = \frac{N!}{\prod_i N_i!}$ gets augmented to $W = \frac{N!}{\prod_i N_i!}\prod_i g_i^{N_i} = N!\prod_i\frac{g_i^{N_i}}{N_i!}$.

For the indistinguishable case, with symmetric wavefunctions:

| $E_1$ | $E_2$ | $\psi$ |
|----------|----------|----------|
| $AA$       |          | $\psi_1(A)\psi_1(A)$
| $A$        | $A$        | $\psi_1(A)\psi_2(B) + \psi_1(B)\psi_2(A)$
|          | $AA$       | $\psi_2(A)\psi_2(A)$


Indistinguishable particles which do not obey to Pauli's exclusion principle, as is the current situation, are refered to as bosons, described by Bose-Einstein's statistics. Now the number of possible distributions is $W_{BE} = \frac{(g+N-1)!}{N!(g-1)!}$. Using the example of $g=2$ and $N=2$, we have that $W_{BE} = \frac{(2+2-1)!}{2!(2-1)!} = \frac{3!}{2!} = 3$. In probability, this is analogous to the stars and bars problem. 

If we go further and draw the indistinguishable case considering Pauli's exclusion principle, then the only possible option given the antisymmetric nature of the wavefunctions is:

| $E_1$ | $E_2$ | $\psi$ |
|----------|----------|----------|
| $A$        | $A$        | $\psi_1(A)\psi_2(B) - \psi_1(B)\psi_2(A)$ |


This table exhibits the only possible states for the current example following Fermi-Dirac's distribution's possible states $W_{FD}$, denoted generically as $W_{FD} = \frac{g!}{N!(g-N)!} = \frac{2!}{2!(2-2)!} = \frac{2!}{2!} = 1$. From probabilistics, this would be the equation for a binomial coefficient.

#### 2.3 Classical Limit

The number of available states for the classical Boltzmann distribution scales faster for the distinguishable particles case, as already seen. As such, a limitation of the classical statistics is applied in order to subject the distinguishable particles to an indistinguishable projection based on Slater's determinant $\sqrt{N!}$, being an $N!$ divisor. In order to proof what has been said, let us apply the limit where $g$ tends to infinity to both indistinguishable case equations shown:

$$W_{BE} = \lim\limits_{g\to\infty} \frac{(g+N-1)!}{N!(g-1)!} = \lim\limits_{g\to\infty} \frac{(g+N-1)(g+N-2)\dots(g+1)g}{N!} = \frac{g^N}{N!}$$

$$W_{FD} = \lim\limits_{g\to\infty} \frac{g!}{N!(g-N)!} = \lim\limits_{g\to\infty} \frac{g(g-1)(g-2)\dots(g-N+1)}{N!} = \frac{g^N}{N!}$$

Both resemble the $W_B$ expression divided by the $N!$ factor. If there are the same $i$ states metioned in the Boltzmann distribution section, then, just like before, we can write: 

$$W_{BE} = \lim\limits_{g\to\infty} \prod_i\frac{(g_i+N_i-1)!}{N_i!(g_i-1)!} = \lim\limits_{g\to\infty} \prod_i\frac{(g_i+N_i-1)(g_i+N_i-2)\dots(g_i+1)g_i}{N_i!} = \prod_i\frac{g_i^{N_i}}{N_i!}$$

$$W_{FD} = \lim\limits_{g\to\infty} \prod_i\frac{g_i!}{N_i!(g_i-N_i)!} = \lim\limits_{g\to\infty} \prod_i\frac{g_i(g_i-1)(g_i-2)\dots(g_i-N_i+1)}{N_i!} = \prod_i\frac{g_i^{N_i}}{N!}$$

#### 3.2 Partition Function and Energy

According to Boltzmann's distribution, the probability of the system to be in some state $i$ is given by $P_i = \frac{N_i}{N} = \frac{\text{e}^{-\beta E_i}}{Z}$, where $Z = \sum_i \text{e}^{-\beta E_i}$ is the (canonical) partition function, a central concept in statistical thermodynamics. Let us have the energy components of the system to be separable, independend of one another. This way $E = E_1 + E_2 +\,...\,+ E_n$, meaning $Z$ is also separable, $Z = Z_1 \cdot Z_2 \cdot\,...\,\cdot Z_n$, since it would be the case that $Z = \sum_i\text{e}^{-\beta \sum_j E_{ij}}$, and connects itself to the classical limit density of states as shows the form $Z = \frac{z^N}{N!}$.

Being the internal energy defined as $U = \sum_i P_i E_i$, we can expand the aforementioned energy sum to $U = \frac{\sum_i E_i\text{e}^{-\beta E_i}}{Z}$. Since $\frac{1}{Z}\frac{\partial Z}{\partial \beta} = \frac{\sum_i E_i\text{e}^{-\beta E_i}}{Z} \implies Z = -\frac{\sum_i E_i\text{e}^{-\beta E_i}}{\frac{\partial \ln\beta}{\partial\beta}} = \frac{-1}{\frac{\partial\ln Z}{\partial\beta\cdot \sum_i E_i\text{e}^{-\beta E_i}}}$, one could write $U = -\frac{\partial \ln Z \cdot \sum_i E_i \text{e}^{-\beta E_i}}{\partial\beta \cdot \sum_i E_i \text{e}^{-\beta E_i}} = -\frac{\partial\ln Z}{\partial \beta}$. Now, tranfering the derivative operation to respect to temperature, $\frac{\text{d}}{\text{d}T} = \frac{\text{d}\beta}{\text{d}T}\frac{\text{d}}{\text{d}\beta} = \frac{\text{d}\frac{1}{k_BT}}{\text{d}T}\frac{\text{d}}{\text{d}\beta} = -\frac{1}{k_BT^2}\frac{\text{d}}{\text{d}\beta} \implies \frac{\text{d}}{\text{d}\beta} = -k_BT^2\frac{\text{d}}{\text{d}T}$. Hence $U = k_BT^2\frac{\partial\ln Z}{\partial T}$.

Analogously, other fundamental relations include: 
$$A = -k_B\ln Z$$

$$U = -T^2\left( \frac{\partial\frac{A}{T}}{\partial T} \right)_{N,V} = kT^2\left( \frac{\partial\ln Z}{\partial T} \right)_{N,V}$$

$$p = -\left( \frac{\partial A}{\partial V} \right)_{N,T} = kt\left( \frac{\partial\frac{A}{T}}{\partial T} \right)_{N,T} = kT\left( \frac{\partial\ln Z}{\partial V}\right)_{N,T}$$

$$\mu = \left( \frac{\partial A}{\partial N} \right)_{V,T} = -kT\left( \frac{\partial\ln Z}{\partial N} \right)_{V,T}$$

$$S = \frac{U-A}{T} = kT\left( \frac{\partial\ln Z}{\partial T} \right)_{N,V}+k\ln Z$$
$$H=U+pV = kT\left( T\left( \frac{\partial\ln Z}{\partial T} \right)_{N,V} + V\left( \frac{\partial\ln Z}{\partial V} \right)_{N,T} \right)$$
$$G = U + pV - TS = -kT\left( \ln Z - V\left( \frac{\partial\ln Z}{\partial V} \right)_{N,T} \right)$$ 
$$C_V = \left( \frac{\partial U}{\partial T} \right)_{N,V} = kT\left( 2\left( \frac{\partial\ln Z}{\partial T} \right)_{N,V} + T\left( \frac{\partial^2\ln Z}{\partial T^2} \right)_{N,V} \right)$$
